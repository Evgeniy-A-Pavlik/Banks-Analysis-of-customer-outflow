## Проверка гипотезы различия дохода между теми клиентами, которые ушли и теми, которые остались

Проверим гипотезу различия среднего дохода между теми клиентами, которые ушли и теми, которые остались

Сформулируем гипотезы:
- $H_0$ — средний доход тех кто ушел и тех кто остался одинаков
- $H_1$ — средний доход тех кто ушел и тех кто остался различается

Применим t-тест

```python
alpha = 0.05

results = st.ttest_ind(df.query('Churn == 0')['estimated_salary'],
                       df.query('Churn == 1')['estimated_salary'], equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- Средний доход тех кто ушел и тех кто остался одинаков

## Формулировка и проверка статистической гипотезы относительно представленных данных

Сформулируем и проверим гипотезу о том, что доля клиентов с кредитной картой ушедших в отток и доля оставшихся клиентов с кредитной картой одинаковы

Сформулируем гипотезы:
- $H_0$ — доля клиентов с кредитной картой ушедших в отток и доля оставшихся клиентов с кредитной картой одинаковы
- $H_1$ — доля клиентов с кредитной картой ушедших в отток и доля оставшихся клиентов с кредитной картой различаются

Применим z-тест

```python
alpha = 0.05

p1 = df.query('Churn == 0 & CreditCard == 1')['userid'].count() / df.query('Churn == 0')['userid'].count()
p2 = df.query('Churn == 1 & CreditCard == 1')['userid'].count() / df.query('Churn == 1')['userid'].count()
p_combined = ((df.query('Churn == 0 & CreditCard == 1')['userid'].count()
               + df.query('Churn == 1 & CreditCard == 1')['userid'].count())
               / (df.query('Churn == 0')['userid'].count() + df.query('Churn == 1')['userid'].count()))
difference = p1 - p2

z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1 / df.query('Churn == 0')['userid'].count() 
                                                                 + 1 / df.query('Churn == 1')['userid'].count()))

distr = st.norm(0, 1)

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными')
```

- Доля клиентов с кредитной картой ушедших в отток и доля оставшихся клиентов с кредитной картой одинаковы
